```python
import requests
import json
from bs4 import BeautifulSoup
import time
import re
from datetime import date
import datetime
import pathlib
import os 



cwd = os.getcwd()
pathlib.Path(cwd + "/DAWN Articles").mkdir(parents=True, exist_ok=True)
today = date.today()
store = []
def main(url):
    with requests.Session() as req:
            soup = ret(url)
            a1 = ""
            for a in soup.find_all('a', href=True):
                    if(a['href'] != a1 and 'https://www.dawn.com/' in a['href'] and '#comments' not in a['href'] and len(a['href']) > 60):
                        sub(a['href'])
                    elif('https://' not in a['href'] and "classifieds" not in a['href'] and len(a['href'])>2):       
                        a['href'] = 'https://www.dawn.com/' + a['href']
                        time.sleep(1)
                        soup2 = ret(a['href'])
                        a2 = ""
                        for b in soup2.find_all('a', href=True):
                            if(b['href'] != a2 and '#comments' not in b['href'] and 'https://www.dawn.com/' in b['href'] and len(b['href']) > 50 and 'https://www.dawn.com/authors' not in b['href']):
                                sub(b['href'])
                            elif('https://www.dawn.com/authors' in b['href'] and len(b['href'])>25):
                                time.sleep(1)
                                soup4 = ret(b['href'])
                                check = soup4.find('ol', {'class':'nav pagination'})
                                for page in check.find_all('a',href=True):
                                    if('https://www.dawn.com' not in page['href']):
                                        page['href'] = 'https://www.dawn.com' + page['href']
                                    time.sleep(1)
                                    soup5 = ret(page['href'])
                                    for c in soup5.find_all('a', href=True):
                                        if(len(c['href'])>60 and 'https://www.dawn.com' in c['href']):
                                            sub(c['href'])
                            a2 = b['href']
                    a1 = a['href']

def sub(url):
    for x in range(3):
        if(url in store):
            break;
        soup3 = ret(url)
        try:
            headline = soup3.find('a', {'class':'story__link'}).get_text()
            auth = soup3.find('div', {"class":'template__header'})                                     
            temp_ls = auth.find_all('a', {"class":'story__byline__link'})
            if(not temp_ls):
                temp_ls = auth.find_all('p', {"class":'story__authors__name'}) 
            if(not temp_ls):
                auth = soup3.find('div', {"class":'template__main'})
                temp_ls = auth.find_all('a', {"class":'story__byline__link'})
            dop = soup3.find('span', {'class':'timestamp--date'})
            if(not dop):
                dop = soup3.find('span', {'class':['timestamp--date','timeago']})
                dop = dop['title']
                dop = re.sub('[^a-zA-Z0-9]', '', dop)
                dop = dop[:-11]
                dop = datetime.datetime.strptime(dop, '%Y%m%d').strftime('%Y-%m-%d')
            else:
                dop = dop.get_text()
                dop = re.sub('[^a-zA-Z0-9]', '', dop)
                if('pm' in dop or 'am' in dop):
                    dop=dop[:-6]
                dop = datetime.datetime.strptime(dop, '%B%d%Y').strftime('%Y-%m-%d')
            data = "Source: DAWN;"
            data += "Link: " + url + ";"
            data += "DateOfExtraction: " + today.strftime("%Y-%m-%d") + ";"
            data += "DateOfPublication: " + dop +";"
            data += "Author:"
            for temp in temp_ls:
                data += temp.get_text() + ","
            data = data[:-1]
            data += ";"
            data += "Title: " + headline + ";"
            data += "Article:" + soup3.find('div', {'class':'story__content'}).get_text()
            f=open(cwd + "/DAWN Articles/" +re.sub('[^a-zA-Z0-9]', '_', headline)+'.txt','w',errors="ignore")
            print(data, file =f)
            f.close()
            store.append(url)
            break;
        except:
            time.sleep(4)
            if(x==2):
                print("Irregular data : ", url)
               
def ret(url):
    with requests.Session() as req:
        try:
            r = req.get(url)
        except requests.exceptions.ConnectionError:
            r.status_code = "Connection refused"
        html_response = r.text
        return BeautifulSoup(html_response,'html.parser')
        

main("https://www.dawn.com")
```
